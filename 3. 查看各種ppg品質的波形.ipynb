```python
import numpy as np
from toolbox_library import Unqualified_PPG_preprocessor
import matplotlib.pyplot as plt

# 指定 .npy 檔案的路徑

# 加載 .npy 檔案
chunk_size = 500
pre = Unqualified_PPG_preprocessor(data_dir='../train', chunk_size=chunk_size)
psycho_ppgs = pre.chunked_psycho
rest_ppgs = pre.chunked_rest

psycho_qualities = np.load('../train/chunked_psycho_quality.npy')
rest_qualities = np.load('../train/chunked_rest_quality.npy')
```

```python
print(len(rest_ppgs))
count = 0
for i in range(len(rest_ppgs)):
    if rest_qualities[i]>=0.96:
        count+=1
print(count)
```

```python
# 生成一個隨機的索引順序
indices = np.arange(rest_qualities.shape[0])
np.random.shuffle(indices)

# 同時打亂 qualities 和 psychos
qualities_shuffled = rest_qualities[indices]
ppgs_shuffled = rest_ppgs[indices]

upper = 0.97
lower = 0.96

count = 0
for i in range(len(indices)):
    quality = qualities_shuffled[i]
    if lower<=qualities_shuffled[i]<=upper:
        print(quality)
        plt.figure(figsize=(5, 3))
        plt.plot(-ppgs_shuffled[i], label='ppg', linestyle='-', linewidth=2)
        plt.legend()
        plt.show()

        count+=1
        if count==30:
            break
```
